```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import rc
```

```python
data = []
f = open("res_GRF.dat","r")
for line in f:
    data.append(np.array(line.split(),dtype=int))
f.close()
```

```python
N = data[0][0]
f = np.zeros((3*N,3*N),dtype=int)
```

```python
fig = plt.figure()
ims = []
for d in data[-100:]:
    B = np.repeat(np.repeat(d[1:N*N+1].reshape(N,N), 2, axis=0), 2, axis=1)
    Nr = d[N*N+1]
    x = d[N*N+2:N*N+2+Nr*2:2]
    y = d[N*N+3:N*N+2+Nr*2:2]
    B[x*2,y*2] = 2
    Nf = d[N*N+2+Nr*2]
    x = d[N*N+3+Nr*2:N*N+3+Nr*2+Nf*2:2]
    y = d[N*N+4+Nr*2:N*N+3+Nr*2+Nf*2:2]
    B[x*2+1,y*2] = 3
    im = plt.imshow(B, cmap="jet", animated=True)
    ims.append([im])
ani = animation.ArtistAnimation(fig, ims, interval=150, blit=True, repeat_delay=0)
plt.show()
```
