```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import rc
```

```python
data = []
f = open("res_BTW.dat","r")
for line in f:
    data.append(np.array(line.split()[1:],dtype=int))
f.close()
```

```python
N = int(np.sqrt(len(data[0])))
```

```python
fig = plt.figure()
ims = []
for d in data[-100:]:
    im = plt.imshow(d.reshape(N,N), cmap="Set1", animated=True,vmax=10,vmin=0)
    ims.append([im])
ani = animation.ArtistAnimation(fig, ims, interval=150, blit=True, repeat_delay=0)
plt.show
```

```python
rc('animation', html='jshtml')
ani
```
