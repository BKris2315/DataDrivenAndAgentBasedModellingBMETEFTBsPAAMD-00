## Agent based modelling

<i>"Computational method that enables a researcher to create, analyze, and experiment with models composed of agents that interact within an environment."</i> - Nigel Gilbert
    
In ABMs agents are usually presenting four important features: 	

 1. <b>Autonomy.</b> There is no global controller indicating the action an agent has to take. Actions are decided according to the programming rules defining the agent;
 2. <b>Social ability.</b> Agents interact among them directly and/or indirectly;
 3. <b>Reactivity.</b> Agents are able to react to stimuli coming from other agents and/or from the environment;
 4. <b>Proactivity.</b> Agents have a goal or goals that they pursue on their initiative.

In specific implementations of ABMs these properties are related to the following characteristics of agents: 
 1. <b>Perception.</b> Agents can perceive their environment and also the behavior of other agents in their vicinity;
 2. <b>Performance.</b> They have a set of behaviors they can perform. Examples are: motion, communication, action, etc;
 3. <b>Memory.</b> They have a memory storing a set of information about a certain number of past states and actions;
 4. <b>Policy.</b> They have a set of rules, heuristics, or strategies that control their response to the present situation taking into account their history.	



<i>"Emergence is generally understood to be a process that leads to the appearance of structure not directly described by the defining constraints and instantaneous forces that control a system. Over time “something new” appears at scales not directly specified by the equations of motion. An emergent feature also cannot be explicitly represented in the initial and boundary conditions. In short, a feature emerges when the underlying system puts some effort into its creation. These observations form an intuitive definition of emergence. For it to be useful, however, one must specify what the “something” is and how it is “new”"</i> - J. P. Crutchfield

## History

### von Neumann self replicating robot idea and cellular automaton

<b> Task: </b> Design an abstract machine which, when run, would replicate itself.

In his design, the self-replicating machine consists of three parts:
 1. <b>description of itself</b>
 2. <b>a universal constructor mechanism</b> that can read any description and construct the machine (sans description) encoded in that description
 3. <b>a universal copy machine</b> that can make copies of any description.
 
After the universal constructor has been used to construct a new machine encoded in the description, the copy machine is used to create a copy of that description, and this copy is passed on to the new machine, resulting in a working replication of the original machine that can keep on reproducing. Some machines will do this backwards, copying the description and then building a machine. Crucially, the self-reproducing machine can evolve by accumulating mutations of the description, not the machine itself, thus gaining the ability to grow in complexity.

### First model simple Conway's game of life

 * Conway experimenting with rules which allows for growth while being difficult to proove
 * Final model is a Turing machine on which all algorithmic calculations are possible
 
### Rules
 1. Square lattice, sites can be empty or occupied by a cell
 2. Any live cell with fewer than two live neighbours dies, as if by underpopulation.
 3. Any live cell with two or three live neighbours lives on to the next generation.
 4. Any live cell with more than three live neighbours dies, as if by overpopulation.
 5. Any dead cell with exactly three live neighbours becomes a live cell, as if by reproduction.

## Lazy bureaucrat model aka sandpile model
<img src="btw.png" width="500px">

<b>Model description</b>
 * Bureaucrats are sitting in a large office with tables in a square lattice arrangement
 * Occasionally the boss comes with a dossier and places it on a random table
 * The bureaucrats do nothing until they have less than 4 dossiers on their table
 * Once a bureaucrat has 4 or more dossiers on its table starts to panic and distributes its dossiers to its 4 neighbors
 * The ones sitting at the windows give also 1 dossier to its neighbors and throw the rest out of the window.
 * <b>No adjustable parameter!!!</b>
